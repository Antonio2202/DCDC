```python
day_dataset = day_dataset[day_dataset['icon'].isin(['cloudy', 'partly-cloudy-day', 'clear-day'])]
night_dataset = night_dataset[night_dataset['icon'].isin(['cloudy', 'partly-cloudy-night', 'clear-night'])]
```

```python
# Verificar las columnas
print(day_dataset[['icon', 'number_tweets']].head())

# Realizar el ANOVA
# Primero, agrupamos los datos por la columna 'icon' y obtenemos las distribuciones de 'number_tweets' por cada grupo.
groups = [day_dataset[day_dataset['icon'] == category]['number_tweets'] 
          for category in day_dataset['icon'].unique()]

# Aplicar la prueba ANOVA
anova_result = stats.f_oneway(*groups)

# Mostrar los resultados de ANOVA
print(f'Valor p del ANOVA: {anova_result.pvalue}')
print(f'Estadístico F: {anova_result.statistic}')

# Interpretación de los resultados
if anova_result.pvalue < 0.05:
    print("Existen diferencias significativas en el número de tweets entre las diferentes condiciones meteorológicas (icon).")
else:
    print("No existen diferencias significativas en el número de tweets entre las diferentes condiciones meteorológicas (icon).")

# Visualización con boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='icon', y='number_tweets', data=day_dataset, palette="Set2")
plt.title('Distribución del número de tweets por condición meteorológica de día')
plt.xlabel('Condición Meteorológica Diurna')
plt.ylabel('Número de Tweets')
plt.show()
```

```python
# Análisis de correlación
# 1. Pearson (correlación lineal)
pearson_corr, pearson_p = pearsonr(df["temp"], df["number_tweets"])
print(f"\nCorrelación de Pearson: {pearson_corr:.2f}, p-valor: {pearson_p:.3e}")

# 2. Spearman (correlación monotónica)
spearman_corr, spearman_p = spearmanr(df["temp"], df["number_tweets"])
print(f"Correlación de Spearman: {spearman_corr:.2f}, p-valor: {spearman_p:.3e}")

# Visualización de la relación
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Gráfico de dispersión
sns.scatterplot(x="temp", y="number_tweets", data=df, alpha=0.6, color="blue")

# Línea de tendencia
sns.regplot(x="temp", y="number_tweets", data=df, scatter=False, color="red")

plt.title("Relación entre temperatura y número de tweets", fontsize=16)
plt.xlabel("Temperatura (°C)", fontsize=14)
plt.ylabel("Número de Tweets", fontsize=14)
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

Night Tweets

```python
from sklearn.feature_extraction import text
from nltk.corpus import stopwords
import nltk

# Descargar las stop words de nltk (solo necesitas hacerlo una vez)
nltk.download('stopwords')

# Obtener las stop words en portugués
stop_words_pt = stopwords.words('portuguese')
# my_additional_stop_words = ['xxx','yyy']
# my_stop_words = text.ENGLISH_STOP_WORDS.union(my_additional_stop_words)


# bag-of-words
tf_vectorizer = CountVectorizer(
    stop_words=stop_words_pt,
    min_df=1,
    max_df=1.0,
    lowercase=True,
    max_features=50000,
    token_pattern='[a-zA-Z0-9]{3,}',
    analyzer = 'word'
)
```

```python
bag_of_words = tf_vectorizer.fit_transform(all_night_tweets_no_urls)
dictionary = tf_vectorizer.get_feature_names_out()
vocabulary = tf_vectorizer.vocabulary_

print("Vocabulary Size: ", len(dictionary))
```

```python
import numpy as np

# Sumar las columnas directamente en la matriz dispersa
s = np.array(bag_of_words.sum(axis=0)).flatten()

# Ordenar las palabras por frecuencia
st = sorted(range(len(s)), key=lambda k: s[k], reverse=True)

# Mostrar las 20 palabras más frecuentes
for i, x in enumerate(st[:20]):
    print(dictionary[x], s[x])
```

```python
from scipy.sparse import csr_matrix
if not isinstance(bag_of_words, csr_matrix):
    bag_of_words = csr_matrix(bag_of_words)
```

```python

topics = 3

alpha = 0.1

beta = 0.05

# Run LDA
lda = LatentDirichletAllocation(
    n_components=topics,
    doc_topic_prior=alpha,
    topic_word_prior=beta,
    max_iter=25,
    learning_method='online',
    evaluate_every=1,
    n_jobs = -1,
    random_state=0,
    verbose=1)
lda.fit(bag_of_words)
```

```python
# Log Likelyhood: Higher the better
print("Log Likelihood: ", lda.score(bag_of_words))

# Perplexity: Lower the better. Perplexity = exp(-1. * log-likelihood per word)
print("Perplexity: ", lda.perplexity(bag_of_words))

# See model parameters
print(lda)
```

```python
no_top_words = 10
no_top_documents = 3

doc_topics = lda.transform(bag_of_words)
topics = lda.components_

print("LDA Topics")
for topic_idx, topic in enumerate(topics):
    print("-"*30)
    print(" Topic ",(topic_idx)," :")
    print("["," | ".join([dictionary[i]
                    for i in topic.argsort()[:-no_top_words - 1:-1]]),"]")
    top_doc_indices = np.argsort( doc_topics[:,topic_idx] )[::-1][0:no_top_documents]
    
```
